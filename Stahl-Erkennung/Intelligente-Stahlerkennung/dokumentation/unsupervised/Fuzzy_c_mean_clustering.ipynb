```python
def fuzzy_c_means_cluster(data, labels_df, name_of_label, number_of_cluster):
    
    # Standardizing the Data
    scaler = FunctionTransformer(np.log1p, validate=True)
    data_scaled = scaler.fit_transform(data)
    
    # Apply fuzzy-c-mean with hard parameters
    cntr, u1, u0, d, jm, p, fpc = fuzz.cluster.cmeans(data_scaled.transpose(), number_of_cluster, 1.1, error=0.005, maxiter=10000, init='PCA')
    
    # Transpose fuzzy matrix u
    u_df = pd.DataFrame(u1.T)
    
    # Create cluster membership overview
    cluster_df = pd.DataFrame(u_df.values.round(6), columns=[f'cluster {i+1}' for i in range(u_df.shape[1])], index=data.index)
    
    # Assign each data point to the cluster with the highest membership
    max_membership_indices = np.argmax(u1, axis=0)

    # Count the number of data points in each cluster
    cluster_counts = np.bincount(max_membership_indices, minlength=u1.shape[0])

    # Add the cluster_counts as the top row of cluster_df
    cluster_df.loc['Insgesamt'] = cluster_counts

    # Add labels to the cluster_df
    cluster_df[name_of_label] = labels_df[name_of_label]

    print("Der folgende Wert gibt an wie klar die Cluster getrennt sind, umso näher an 1 desto klarer sind sie getrennt:", fpc.round(5))
    
    return cluster_df
```

```python
def fuzzy_c_means_clustering(data,number_of_clusters):
    # Extracting features and class labels
    features = data.iloc[:, :-1].values.T
    labels = data.iloc[:, -1]

    # Perform fuzzy c-means clustering
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
        features, number_of_clusters, 1.5, error=0.005, maxiter=10000, init=None
    )

    # Determining the cluster to which each data point belongs
    cluster_membership = np.argmax(u, axis=0)

    # Creating a dataframe to count occurrences
    cluster_counts = pd.DataFrame(0, index=np.unique(labels), columns=np.arange(number_of_clusters))

    # Counting occurrences
    for label, cluster in zip(labels, cluster_membership):
        cluster_counts.loc[label, cluster] += 1

    return cluster_counts
```
